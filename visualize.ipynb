```python
# 查看用 expert 训练得到的效果
# 对比是否用 Expert
# 2019/7/1 13：00

import torch
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

import warnings
warnings.filterwarnings('ignore')

plt.figure(figsize=(15,8))
seeds = [3]

env_names = ['HalfCheetah-v2', 'Hopper-v2', 'Walker2d-v2']
for i in range(len(env_names)):
    env_name = env_names[i]
    for seed in seeds:
        file1 = './results/'+env_name+'/ExpertDDPG_'+env_name+'_%d.npz'%seed
        file0 = './results/'+env_name+'/ExpertDDPG_'+env_name+'_%d_without_expert.npz'%seed
        
        plt.subplot(2,3,i+1).set_title(env_name)
        f = np.load(file1)
        steps = f['value_step']
        value_gd = np.mean(f['value_true'],1)
        value_pred = np.mean(f['value_pred'],1)
        episode_reward = f['episode_reward']
        plt.plot(steps/1000, value_gd, 'r--',label='True DDPG')
        plt.plot(steps/1000, value_pred, 'r-',label='Ours DDPG')
        plt.subplot(2,3,i+4)
        plt.plot(f['episode_reward'], 'r', label='WithExpert')
        plt.xlabel('Episode')
        
        plt.subplot(2,3,i+1)
        f = np.load(file0)
        steps = f['value_step']
        value_gd = np.mean(f['value_true'],1)
        value_pred = np.mean(f['value_pred'],1)
        episode_reward = f['episode_reward']
        plt.plot(steps/1000, value_gd, 'b--',label='True DDPG(NoExpert)')
        plt.plot(steps/1000, value_pred, 'b-',label='Ours DDPG(NoExpert)')
        
        plt.axvline(f['expert_timesteps']/1000,linestyle='--')
        plt.axhline(f['value_expert'],linestyle='--')
        plt.xlabel('Time steps(1e3)')
        plt.ylabel('Average Value')
        plt.legend(loc=0)
        
        plt.subplot(2,3,i+4)
        plt.plot(f['episode_reward'], 'b', label='NoExpert')
        plt.legend(loc=0)
        if i == 0:
            plt.ylabel('Episode Reward')
```
